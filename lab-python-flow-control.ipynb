```python
#2 Changes:

products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {}

for p in products:
    inventory[p] = int(input(f"Enter the inventory count for {p}: "))

print("Inventory:",inventory)


customer_orders = set()

#Nuevas condiciones
while True:
    deseo = input("Ingrese el nombre del producto que desea ordenar: ").strip().lower()
    if deseo in products:
        customer_orders.add(deseo)
        o = input("¿Quieres añadir otro producto? (yes/no): ").strip().lower()
        if o == "no":
            break
    else:
        print(f"El producto '{deseo}' no está disponible")
        continue
    

print("Customer orders:", customer_orders)

total_products_ordered = len(customer_orders)
percentage_ordered = (total_products_ordered / len(products)) * 100
order_status = (total_products_ordered, percentage_ordered)
print(order_status)

print("Order Statistics:")
print("Total Products Ordered:", order_status[0])
print("Percentage of Products Ordered:", f"{order_status[1]:.2f}%")

# 3) Actualizar inventario SOLO para los productos pedidos (1 unidad por producto distinto)
for product in customer_orders:
    inventory[product] = max(inventory[product] - 1, 0)

print("Updated Inventory:")
for product in products:
    print(f"{product}: {inventory[product]}")
```
